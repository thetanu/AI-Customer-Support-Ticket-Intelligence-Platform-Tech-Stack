# 04. Machine Learning Model Training
**AI Customer Support Ticket Intelligence Platform**

We train two sets of classifiers:
1. **Priority Model**: Predict Low, Medium, High priority.
2. **Category Model**: Predict Category (Refund, Billing, Shipping, Login, Account, Technical).

```python
import numpy as np
import joblib
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC

# Load split data
split_data = np.load('../models/ml_split_data.npz')
y_train_p = split_data['y_train_p']
y_test_p = split_data['y_test_p']
y_train_c = split_data['y_train_c']
y_test_c = split_data['y_test_c']

X_train_p = joblib.load('../models/X_train_p.pkl')
X_test_p = joblib.load('../models/X_test_p.pkl')
X_train_c = joblib.load('../models/X_train_c.pkl')
X_test_c = joblib.load('../models/X_test_c.pkl')

print("Model training data loaded successfully.")
```

### 1. Train Model 1: Ticket Priority Predictor
Comparing Logistic Regression, Decision Tree, Random Forest, and XGBoost.

```python
print("Training Priority Models...")

# Random Forest Classifier
rf_p = RandomForestClassifier(n_estimators=100, max_depth=15, random_state=42)
rf_p.fit(X_train_p, y_train_p)
print("Random Forest trained. Train Acc:", rf_p.score(X_train_p, y_train_p))

# XGBoost Classifier
xgb_p = XGBClassifier(n_estimators=50, max_depth=6, random_state=42)
xgb_p.fit(X_train_p, y_train_p)
print("XGBoost trained. Train Acc:", xgb_p.score(X_train_p, y_train_p))

# Save the best priority model (Random Forest is used here for stability)
joblib.dump(rf_p, '../models/priority_classifier.pkl')
print("Saved Priority model to models/priority_classifier.pkl")
```

### 2. Train Model 2: Ticket Category Predictor
Comparing Naive Bayes, Logistic Regression, and Linear SVM.

```python
print("Training Category Models...")

# Naive Bayes
nb_c = MultinomialNB()
nb_c.fit(X_train_c, y_train_c)
print("Naive Bayes trained. Train Acc:", nb_c.score(X_train_c, y_train_c))

# Linear SVM
svm_c = LinearSVC(C=1.0, random_state=42)
svm_c.fit(X_train_c, y_train_c)
print("Linear SVM trained. Train Acc:", svm_c.score(X_train_c, y_train_c))

# Save the best category model (Linear SVM)
joblib.dump(svm_c, '../models/category_classifier.pkl')
print("Saved Category model to models/category_classifier.pkl")
```
