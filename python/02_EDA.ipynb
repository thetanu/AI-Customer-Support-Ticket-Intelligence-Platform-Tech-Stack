# 02. Exploratory Data Analysis (EDA)
**AI Customer Support Ticket Intelligence Platform**

In this notebook, we visualize operational KPIs, ticket distribution patterns, agent response times, customer satisfaction levels, outlier limits, and generate text visualizations.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud

sns.set_theme(style="whitegrid")
cleaned_csv_path = '../dataset/cleaned_customer_support_tickets.csv'

df = pd.read_csv(cleaned_csv_path)
print(f"Dataset loaded. Columns: {df.columns.tolist()}")
```

### 1. Ticket Volume Distributions (Category, Priority, Status)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.countplot(data=df, x='Ticket_Category', ax=axes[0], palette='viridis', order=df['Ticket_Category'].value_counts().index)
axes[0].set_title('Tickets by Category')
axes[0].tick_params(axis='x', rotation=45)

sns.countplot(data=df, x='Priority', ax=axes[1], palette='crest', order=['Low', 'Medium', 'High'])
axes[1].set_title('Tickets by Priority')

sns.countplot(data=df, x='Status', ax=axes[2], palette='magma')
axes[2].set_title('Tickets by Status')
axes[2].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

### 2. Resolution Time Distribution & Outlier Detection

```python
closed_tickets = df[df['Status'] == 'Closed'].copy()

plt.figure(figsize=(10, 5))
sns.histplot(data=closed_tickets, x='Resolution_Time', kde=True, color='purple', bins=30)
plt.title('Distribution of Resolution Time (Hours) for Closed Tickets')
plt.xlabel('Hours to Resolve')
plt.show()

plt.figure(figsize=(10, 4))
sns.boxplot(data=closed_tickets, x='Resolution_Time', color='lightblue')
plt.title('Boxplot of Resolution Time (Hours)')
plt.show()
```

### 3. Customer Demographics & Customer Satisfaction Rating (CSAT)
* Age group distributions
* Satisfaction scores (excluding unrated tickets marked as -1)

```python
rated_tickets = df[df['Customer_Satisfaction_Rating'] >= 1]

plt.figure(figsize=(8, 4))
sns.countplot(data=rated_tickets, x='Customer_Satisfaction_Rating', palette='coolwarm')
plt.title('Customer Satisfaction Rating (CSAT) Distribution')
plt.xlabel('Rating (1-5)')
plt.show()

print(f"Overall Average CSAT Score: {rated_tickets['Customer_Satisfaction_Rating'].mean():.2f}")
```

### 4. Text Analysis & Word Cloud
Extracting common terms from ticket subject descriptions.

```python
text_data = " ".join(str(desc) for desc in df['Ticket_Description'].dropna())

wordcloud = WordCloud(width=800, height=400, background_color='white',
                      colormap='viridis', max_words=100).generate(text_data)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Support Tickets Descriptions', fontsize=16)
plt.show()
```

### 5. Interactive Channel Distribution (Plotly)

```python
channel_counts = df['Channel'].value_counts().reset_index()
channel_counts.columns = ['Channel', 'Count']

fig = px.pie(channel_counts, values='Count', names='Channel', 
             title='Interactive Ticket Channels Share', 
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.show()
```
