# 05. Model Evaluation & Metrics Report
**AI Customer Support Ticket Intelligence Platform**

Evaluating classification models on the unseen holdout test set using Accuracy, Precision, Recall, F1 Score, and plotting confusion matrices.

```python
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load data & encoders
split_data = np.load('../models/ml_split_data.npz')
y_test_p = split_data['y_test_p']
y_test_c = split_data['y_test_c']

X_test_p = joblib.load('../models/X_test_p.pkl')
X_test_c = joblib.load('../models/X_test_c.pkl')

le_priority = joblib.load('../models/priority_encoder.pkl')
le_category = joblib.load('../models/category_encoder.pkl')

model_p = joblib.load('../models/priority_classifier.pkl')
model_c = joblib.load('../models/category_classifier.pkl')
```

### 1. Evaluate Model 1: Ticket Priority Classifier

```python
y_pred_p = model_p.predict(X_test_p)
print("=== Ticket Priority Classification Report ===")
print(classification_report(y_test_p, y_pred_p, target_names=le_priority.classes_))

# Confusion Matrix Plot
cm_p = confusion_matrix(y_test_p, y_pred_p)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_p, annot=True, fmt='d', cmap='Blues', 
            xticklabels=le_priority.classes_, yticklabels=le_priority.classes_)
plt.title('Priority Model Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

### 2. Evaluate Model 2: Ticket Category Classifier

```python
y_pred_c = model_c.predict(X_test_c)
print("=== Ticket Category Classification Report ===")
print(classification_report(y_test_c, y_pred_c, target_names=le_category.classes_))

# Confusion Matrix Plot
cm_c = confusion_matrix(y_test_c, y_pred_c)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_c, annot=True, fmt='d', cmap='Purples', 
            xticklabels=le_category.classes_, yticklabels=le_category.classes_)
plt.title('Category Model Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
