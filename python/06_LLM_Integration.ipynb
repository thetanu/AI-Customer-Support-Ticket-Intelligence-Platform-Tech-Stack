# 06. Generative AI & Gemini API Integration
**AI Customer Support Ticket Intelligence Platform**

We integrate the Google Gemini API to build generative intelligence layers for agents:
1. **AI Ticket Summarization**: Condensing long user descriptions.
2. **AI Response Drafting**: Automatically creating email drafts based on ticket category.
3. **AI Sentiment Extraction**: Evaluating tone and satisfaction levels.

```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

# Load API Key from .env
load_dotenv(dotenv_path='../.env')
api_key = os.getenv("GEMINI_API_KEY")

if api_key:
    genai.configure(api_key=api_key)
    print("Gemini API configured successfully.")
else:
    print("API Key not found. Please add GEMINI_API_KEY to your .env file.")
```

```python
# Example Ticket Description
sample_desc = (
    "I bought a GoPro Hero action camera last week. Today, I tried to connect it to my MacBook Pro "
    "via USB-C, but it is not showing up as an external drive. I tried three different cables, "
    "restarted both the MacBook and GoPro, but nothing worked. I also tried updating the GoPro firmware, "
    "but I cannot copy the firmware update files because my laptop doesn't detect the camera. "
    "This is frustrating since I have an upcoming trip tomorrow and need to transfer my recordings!"
)

sample_subject = "GoPro Hero USB Connection Error with MacBook Pro"
```

### 1. Generate AI Summary of Ticket

```python
def generate_summary(description):
    if not api_key:
        return "Mock Summary: Customer's GoPro Hero camera is not detected by MacBook Pro over USB-C."
    
    model = genai.GenerativeModel('gemini-1.5-flash')
    prompt = f"Summarize this customer support ticket description in exactly one clear sentence:\n\n{description}"
    response = model.generate_content(prompt)
    return response.text.strip()

print("AI Generated Summary:")
print(generate_summary(sample_desc))
```

### 2. Generate Sentiment and Tone Analysis

```python
def extract_sentiment(description):
    if not api_key:
        return "Sentiment: Angry, Tone: Frustrated"
    
    model = genai.GenerativeModel('gemini-1.5-flash')
    prompt = (
        f"Identify the customer sentiment (Positive, Neutral, Negative) and primary tone (e.g. Frustrated, Confused, Polite) "
        f"for this ticket description:\n\n{description}\n\nFormat output as JSON: {{\"sentiment\": \"...\", \"tone\": \"...\"}}"
    )
    response = model.generate_content(prompt)
    return response.text.strip()

print("AI Sentiment Analysis:")
print(extract_sentiment(sample_desc))
```

### 3. Generate Auto-draft Response

```python
def draft_response(subject, description):
    if not api_key:
        return "Mock Draft: Dear Customer, we are sorry for the issue. Please check your camera settings..."
    
    model = genai.GenerativeModel('gemini-1.5-flash')
    prompt = (
        f"Draft a professional customer support email response to this issue.\n"
        f"Subject: {subject}\n"
        f"Issue: {description}\n\n"
        f"The email should thank the customer, acknowledge their frustration, offer a brief troubleshooting step, "
        f"and sign off as 'AI Customer Intelligence Copilot'."
    )
    response = model.generate_content(prompt)
    return response.text.strip()

print("AI Drafted Response:")
print(draft_response(sample_subject, sample_desc))
```
